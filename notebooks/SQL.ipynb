```python
# For analysing the data, I used SQL library on Python

# Pandas & SQLite3 libraries are imported
# After exploring the data on python
# I continue analysing data via importing SQL on Python using Anaconda
import pandas as pd
import sqlite3

# Loading the dataset again
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)

# Displaying the first few rows of the dataset
#to be sure the data is read without an error
print(titanic.head())


# Creating & Connecting to SQLite database
conn = sqlite3.connect('titanic.db')

# Creating a cursor object
cursor = conn.cursor()


# Importing data from pandas DataFrame to SQLite table
titanic.to_sql('titanic', conn, index=False, if_exists='replace')

# Calculating the Overall Survival Rate for the objectives & questions
query = 'SELECT AVG(Survived) AS OverallSurvivalRate FROM titanic;'
result = cursor.execute(query).fetchall()
# Displayin the Overall Survival Rate
print("")
print('Overall Survival Rate:')
print(result)
print("")

# Calculating the Survival Rates by Gender for the objectives & questions 
query = 'SELECT Sex, AVG(Survived) AS SurvivalRate FROM titanic GROUP BY Sex;'
result = cursor.execute(query).fetchall()
# Displaying the Survival Rates by Gender
print('Survival Rate by Gender:')
print(result)
print("")

# Calculating the Survival Rates by Class for the objectives & questions
query = 'SELECT Pclass, AVG(Survived) AS SurvivalRate FROM titanic GROUP BY Pclass;'
result = cursor.execute(query).fetchall()
# Displaying the Survival Rates by Class
print('Survival Rate by Class:')
print(result)
print("")

# Calculating the Survival Rates by Age Group for the objectives & questions
query = "SELECT CASE WHEN Age BETWEEN 0 AND 18 THEN '0-18' WHEN Age BETWEEN 19 AND 35 THEN '19-35' WHEN Age BETWEEN 36 AND 50 THEN '36-50' WHEN Age BETWEEN 51 AND 80 THEN '51-80' ELSE 'Unknown' END AS AgeGroup, AVG(Survived) AS SurvivalRate FROM titanic WHERE Age IS NOT NULL GROUP BY AgeGroup;"

result = cursor.execute(query).fetchall()
# Displaying the Survival Rates by Age Group
print('Survival Rate by Age Group:')
print(result)
print("")

# Indicating the Correlation between Fare and Survival or the objectives & questions
query = 'SELECT Fare, AVG(Survived) AS SurvivalRate FROM titanic GROUP BY Fare;'
result = cursor.execute(query).fetchall()
# Displaying the Correlation between Fare and Survival
print('Correlation between Fare and Survival:')
for fare, survival_rate in result:
    print(f"Fare: {fare}, Survival Rate: {survival_rate}")

# Closing the cursor and the connection
cursor.close()
conn.close()
```
