```python
# For collecting and exploring the data, I used Python via Anaconda

# Importing Pandas, Matplotlib & Seaborn for exploring the data
# I choose to use Python via Anaconda
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from the original URL
# The dataset also available in the repo as titanic.csv (also titanic.db)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)

# Displaying the first few rows of the dataset to be sure
print(titanic.head())

# Checking for any missing values
print("\nMissing Values:")
print(titanic.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(titanic.describe())

# Exploring the data visually via Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=False, color='skyblue')
plt.title('Distribution of Passengers\' Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Exploring the data visually via Survival by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=titanic, palette='viridis')
plt.title('Survival Count by Gender')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()

# Calculating the overall survival rate
overall_survival_rate = titanic['Survived'].mean()
print("\nOverall Survival Rate:", overall_survival_rate)

# Visualizing the survival rates by gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=titanic, palette='muted')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Visualizing the survival rates by class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic, palette='pastel')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Visualizing survival rates by age group
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=[0, 18, 35, 50, 80], labels=['0-18', '19-35', '36-50', '51-80'])
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='Survived', data=titanic, palette='dark')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

# Visualizing the distribution of fares
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Fare'], bins=30, kde=False, color='coral')
plt.title('Distribution of Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
